```python
import xrpl
import nest_asyncio
nest_asyncio.apply()

# Connect to Testnet
client = xrpl.clients.JsonRpcClient("https://s.altnet.rippletest.net:51234")

# YOUR wallet details (from the faucet in Step 1)
my_secret = "sEdVgdqhp8be26D8P4DAVPEDfpvHwS6"  # ← paste your secret between the quotes

# Load your wallet from your secret key
my_wallet = xrpl.wallet.Wallet.from_seed(my_secret)

# Destination wallet (this is a practice address - safe to send to)
destination_address = "rPT1Sjq2YGrBMTttX4GZHjKu9dyfzbpAYe"

# Build the payment transaction
# 10 XRP = 10,000,000 drops
payment = xrpl.models.transactions.Payment(
    account=my_wallet.classic_address,
    amount="10000000",
    destination=destination_address,
)

# Sign and submit to the blockchain
response = xrpl.transaction.submit_and_wait(payment, client, my_wallet)

# Show the result
print("Transaction result:", response.result["meta"]["TransactionResult"])
print("Transaction hash:", response.result["hash"])
print(f"Explorer link: https://testnet.xrpscan.com/tx/{response.result['hash']}")
```
